```python
!pip install qiskit qiskit-aer


# Qiskit 2.x Example ‚Äì OpenQASM Demonstration

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import qiskit.qasm3 as qasm3  # For OpenQASM 3 export


# Create a simple 2-qubit quantum circuit
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()


print("Quantum Circuit:")
print(qc.draw())


# Convert the circuit to OpenQASM 3 format
qasm_code = qasm3.dumps(qc)
print("\nGenerated OpenQASM 3 Code:\n")
print(qasm_code)


# Optional: Simulate the circuit
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
result = simulator.run(compiled_circuit).result()
counts = result.get_counts()
print("\nSimulation Result:")
print(counts)
```

```python
# ===============================================================
# üß† OpenQASM Demonstration using Qiskit 2.x
# ===============================================================

!pip install -q qiskit qiskit-aer matplotlib

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import qiskit.qasm3 as qasm3
import matplotlib.pyplot as plt
from qiskit.visualization import plot_histogram

# ===============================================================
# üß© Task 1 ‚Äî Modify the circuit to have 3 qubits and apply a different gate sequence
# ===============================================================
print("\n=== üß© Task 1: Create 3-Qubit Quantum Circuit with Different Gate Sequence ===")

qc3 = QuantumCircuit(3)
qc3.h(0)          # Hadamard on qubit 0
qc3.cx(0, 1)      # CNOT with control 0, target 1
qc3.cz(1, 2)      # Controlled-Z with control 1, target 2
qc3.x(2)          # Pauli-X on qubit 2
qc3.measure_all() # Measure all qubits

print("\n3-Qubit Circuit:")
print(qc3.draw("text"))

# ===============================================================
# üß© Task 2 ‚Äî Generate and display the QASM code for the new circuit
# ===============================================================
print("\n=== üß© Task 2: Generated OpenQASM 3 Code ===")

qasm_code3 = qasm3.dumps(qc3)
print(qasm_code3)

# ===============================================================
# üß© Task 3 ‚Äî Identify how Hadamard and CNOT operations are represented in QASM syntax
# ===============================================================
print("\n=== üß© Task 3: QASM Representation of Gates ===")

print("""
üß† In OpenQASM 3 syntax:
- Hadamard gate ‚Üí 'h q[0];'
- CNOT (Controlled-X) gate ‚Üí 'cx q[0], q[1];'
- Controlled-Z gate ‚Üí 'cz q[1], q[2];'
- Pauli-X gate ‚Üí 'x q[2];'
""")

# ===============================================================
# üß© Task 4 ‚Äî Add a barrier and observe how it appears in QASM
# ===============================================================
print("\n=== üß© Task 4: Add Barrier and Observe QASM ===")

qc3_barrier = QuantumCircuit(3)
qc3_barrier.h(0)
qc3_barrier.cx(0, 1)
qc3_barrier.barrier()  # Barrier added here
qc3_barrier.cz(1, 2)
qc3_barrier.x(2)
qc3_barrier.measure_all()

print("\nCircuit with Barrier:")
print(qc3_barrier.draw("text"))

qasm_barrier = qasm3.dumps(qc3_barrier)
print("\nQASM Code with Barrier:\n")
print(qasm_barrier)

print("""
üß© Note: The barrier appears in QASM as:
    barrier q[0], q[1], q[2];
It visually separates circuit segments to preserve logical grouping.
""")

# ===============================================================
# üß© Task 5 ‚Äî Run the modified circuit and compare simulation results
# ===============================================================
print("\n=== üß© Task 5: Simulate and Compare Results ===")

sim = AerSimulator()
compiled_original = transpile(qc3, sim)
compiled_barrier = transpile(qc3_barrier, sim)

result_original = sim.run(compiled_original, shots=1024).result()
result_barrier = sim.run(compiled_barrier, shots=1024).result()

counts_original = result_original.get_counts()
counts_barrier = result_barrier.get_counts()

print("\nOriginal Circuit Simulation Results:", counts_original)
print("Barrier Circuit Simulation Results:", counts_barrier)

plot_histogram([counts_original, counts_barrier],
               legend=["Without Barrier", "With Barrier"],
               title="Task 5: Comparison of Simulation Results")
```
